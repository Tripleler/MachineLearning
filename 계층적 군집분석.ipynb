```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
def dist(a, b):
    return ((b[1] - a[1])**2 + (b[0] - a[0])**2)**0.5
```

```python
a = np.array([1, 4])
b = np.array([2, 1])
c = np.array([4, 6])
d = np.array([4, 3])
e = np.array([5, 1])
X = np.array([a, b, c, d, e])
labels = ['a', 'b', 'c', 'd', 'e']
```

```python
plt.figure(figsize=(7, 7))
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.scatter(X[:, 0], X[:, 1])
for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

# 최단연결법(Single Linkage)

```python
linkage(X, 'single')
```

```python
plt.figure(figsize=(7, 7))
plt.gca().add_patch(
    patches.Ellipse((4.5, 2), 4, 1, angle=-60, edgecolor='blue', fill=False))
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.scatter(X[:, 0], X[:, 1])
for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(label,
                 xy=(x, y),
                 xytext=(-3, 3),
                 textcoords='offset points',
                 ha='right',
                 va='bottom')
plt.plot(a, d, color="black")
plt.show()
```

```python
plt.figure(figsize=(7, 7))
dendrogram(linkage(X, 'single'),
           orientation='top',
           distance_sort='descending',
           labels=labels,
           show_leaf_counts=True)
plt.show()
```

```python
plt.figure(figsize=(7, 7))
dendrogram(linkage(X, 'complete'),
           orientation='top',
           distance_sort='descending',
           labels=labels,
           show_leaf_counts=True)
plt.show()
```

# 와드연결법(Ward linkage)

```python
linkage(X, 'ward')
```

```python
plt.figure(figsize=(7, 7))
plt.gca().add_patch(
    patches.Ellipse((4.5, 2), 4, 1, angle=-65, edgecolor='blue', fill=False))
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.scatter(X[:, 0], X[:, 1])
for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(label,
                 xy=(x, y),
                 xytext=(-3, 3),
                 textcoords='offset points',
                 ha='right',
                 va='bottom')
plt.plot([4.5, 1], [2, 4], color="black")
plt.plot([4, 5], [3, 1], color="red")
plt.show()
```

```python
plt.figure(figsize=(7, 7))
dendrogram(linkage(X, 'ward'),
           orientation='top',
           distance_sort='descending',
           labels=labels,
           show_leaf_counts=True)
plt.show()
```

```python
((dist(b, (b + d + e) / 3)**2 + dist(d, (b + d + e) / 3)**2 + dist(e, (b + d + e) / 3)**2 -
  (dist(d, (d + e) / 2)**2 + dist(e, (d + e) / 2)**2)) * 2)**0.5
```

```python
((dist(a, (a + b + c + d + e) / 5)**2 + dist(b, (a + b + c + d + e) / 5)**2 + 
  dist(c, (a + b + c + d + e) / 5)**2 + dist(d, (a + b + c + d + e) / 5)**2 +
  dist(e, (a + b + c + d + e) / 5)**2 - 
  (dist(a, (a + c) / 2)**2 + dist(c, (a + c) / 2)**2) -
  (dist(b, (b + d + e) / 3)**2 + dist(d, (b + d + e) / 3)**2 + dist(e, (b + d + e) / 3)**2)) * 2)**0.5
```
